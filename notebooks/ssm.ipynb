# State-space models

This is an example of a state-space model



```scala211
import breeze.linalg.DenseVector
import org.ruivieira.science.ssm.univariate.{StateGenerator, UnivariateGenerator, UnivariateStructure}

val structure = UnivariateStructure.createLocallyConstant(W = 1.5)

val states = StateGenerator.states(nobs = 1000,
  structure = structure,
  state0 = DenseVector[Double](0.0))

val observations = UnivariateGenerator.gaussian(states = states,
  structure = structure,
  V = 1.0)

```
